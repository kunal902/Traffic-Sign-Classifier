### Pre-process the Data Set (normalization, grayscale, etc.)

Minimally, the image data should be normalized so that the data has mean zero and equal variance. For image data, `(pixel - 128)/ 128` is a quick way to approximately normalize the data and can be used in this project. 

Other pre-processing steps are optional. You can try different techniques to see if it improves performance. 

Use the code cell (or multiple code cells, if necessary) to implement the first step of your project.

````python
### Preprocess the data here. It is required to normalize the data. Other preprocessing steps could include 
### converting to grayscale, etc.
### Feel free to use as many code cells as needed.

from sklearn.utils import shuffle

# convert the data set to grayscale 
def convert_rgb_to_gray(img):
    gray = np.dot(img, [0.299, 0.587, 0.114])
    return gray.reshape(len(img), 32, 32, 1)

# apply min-max normalization to the data set
def min_max_normalization(img):
    minimum = np.min(img)
    maximum = np.max(img)
    return (img - minimum) * 255 / (maximum - minimum)

# apply standardization 
def standardization(img):
    img -= np.mean(img)
    img /= np.std(img)
    return img

# preprocessing pipeline
def preprocessing_pipeline(img):
    img = convert_rgb_to_gray(img)
    img = min_max_normalization(img)
    return standardization(img)

X_net_train = preprocessing_pipeline(X_train)
y_net_train = y_train

X_net_valid = preprocessing_pipeline(X_valid)
y_net_valid = y_valid

X_net_test = preprocessing_pipeline(X_test)
y_net_test = y_test

print(X_net_train[0].shape)
````

### Model Architecture

````python
### Define your architecture here.
### Feel free to use as many code cells as needed.

from tensorflow.contrib.layers import flatten

def LeNet(x):    
    # Layer 1: Convolutional. Input = 32x32x1. Output = 28x28x6.
    weights_layer1 = tf.Variable(tf.truncated_normal(shape=(5, 5, 1, 6), mean=mu, stddev=sigma))
    bias_layer1 = tf.Variable(tf.zeros(6))
    conv1 = tf.nn.conv2d(x, weights_layer1, strides=[1, 1, 1 , 1], padding='VALID') + bias_layer1

    # Activation.
    conv1 = tf.nn.relu(conv1)

    # Pooling. Input = 28x28x6. Output = 14x14x6.
    conv1 = tf.nn.max_pool(conv1, ksize=[1,2,2,1], strides=[1,2,2,1], padding='VALID')

    # Layer 2: Convolutional. Output = 10x10x16.
    weights_layer2 = tf.Variable(tf.truncated_normal(shape=(5, 5, 6, 16), mean=mu, stddev=sigma))
    bias_layer2 = tf.Variable(tf.zeros(16))
    conv2 = tf.nn.conv2d(conv1, weights_layer2, strides=[1, 1, 1 , 1], padding='VALID') + bias_layer2
    
    # Activation.
    conv2 = tf.nn.relu(conv2)

    # Pooling. Input = 10x10x16. Output = 5x5x16.
    conv2 = tf.nn.max_pool(conv2, ksize=[1,2,2,1], strides=[1,2,2,1], padding='VALID')

    # Flatten. Input = 5x5x16. Output = 400.
    flattened_layer = flatten(conv2)
    
    # Layer 3: Fully Connected. Input = 400. Output = 120.
    c_weight1 = tf.Variable(tf.truncated_normal(shape=(400, 120), mean=mu, stddev=sigma))
    c_bias1 = tf.Variable(tf.zeros(120))
    c_layer1 = tf.matmul(flattened_layer, c_weight1) + c_bias1
    
    
    # Activation.
    c_layer1 = tf.nn.relu(c_layer1)
    
    # Dropout for regularization
    c_layer1 = tf.nn.dropout(c_layer1, keep_prob)

    # Layer 4: Fully Connected. Input = 120. Output = 84.
    c_weight2 = tf.Variable(tf.truncated_normal(shape=(120, 84), mean=mu, stddev=sigma))
    c_bias2 = tf.Variable(tf.zeros(84))
    c_layer2 = tf.matmul(c_layer1, c_weight2) + c_bias2
    
    # Activation.
    c_layer2 = tf.nn.relu(c_layer2)
    
    # Dropout for regularization
    c_layer2 = tf.nn.dropout(c_layer2, keep_prob)

    # Layer 5: Fully Connected. Input = 84. Output = 43.
    c_weight3 = tf.Variable(tf.truncated_normal(shape=(84, 43), mean=mu, stddev=sigma))
    c_bias3 = tf.Variable(tf.zeros(43))
    logits = tf.matmul(c_layer2, c_weight3) + c_bias3
    
    return logits
````

## Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
### Train your model here.
### Calculate and report the accuracy on the training and validation set.
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.
### Feel free to use as many code cells as needed.

import tensorflow as tf

tf.reset_default_graph()

#declare variables
x = tf.placeholder(tf.float32, (None, 32, 32, 1))
y = tf.placeholder(tf.int32, (None))
keep_prob = tf.placeholder(tf.float32)
one_hot_y = tf.one_hot(y, 43)
save_file = "./model/model.ckpt"
````

````python
#Hyperparameters
EPOCHS = 200
BATCH_SIZE = 128
rate = 0.001
mu = 0
sigma = 0.1
````

````python
# training pipeline
logits = LeNet(x)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation = optimizer.minimize(loss_operation)
````

````python
#Model Evaluation
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 1.0})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
````

````python
#Train the model
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_net_train)
    
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_net_train, y_net_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 0.8})
            
        validation_accuracy = evaluate(X_net_valid, y_net_valid)
        print("EPOCH {} ...".format(i+1))
        print("Validation Accuracy = {:.3f}".format(validation_accuracy))
        print()
        
    saver.save(sess, save_file)
    print("Model saved")
````

````python
#calculating validation, train, test set accuracy
with tf.Session() as sess:
    saver.restore(sess, save_file)
    print()
    print("Training accuracy = {:.3f}".format(evaluate(X_net_train, y_net_train)))
    print("Validation accuracy = {:.3f}".format(evaluate(X_net_valid, y_net_valid)))
    print("Test accuracy = {:.3f}".format(evaluate(X_net_test, y_net_test)))
````
